```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv("tmdb_5000_movies.csv")
credits = pd.read_csv("tmdb_5000_credits.csv")
```

```python
movies.head(1)
```

```python
credits.head(1)
```

```python
# Merging both datasets
movies = movies.merge(credits, on="title")
```

```python
movies.head(1)
```

```python
# Filtering out Important Columns from the dataset to be included further
movies = movies[['movie_id',  'title', 'overview','genres', 'keywords', 'cast', 'crew']]
```

```python
movies.head()
```

```python
# Extracting names from the list of dictionaries in genres, keywords, cast and crew cols.

# For genres and keywords
import ast
def extract_name(obj):
    N = []
    for i in ast.literal_eval(obj):
        N.append(i['name'].replace(" ", ""))
    return N

movies['genres'] = movies['genres'].apply(extract_name)
movies['keywords'] = movies['keywords'].apply(extract_name)
```

```python
# for cast
def extract_cast_name(obj):
    N = []
    count = 0
    for i in ast.literal_eval(obj):
        if count != 3:
            N.append(i['name'].replace(" ", ""))
            count += 1
        else:
            break
    return N

movies['cast'] = movies['cast'].apply(extract_cast_name)
```

```python
# for crew 
def fetch_director(obj):
    N = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            N.append(i['name'].replace(" ", ""))
            break
    return N

movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.dropna(inplace=True)
```

```python
movies.head()
```

```python
# converting Overview col to list of words
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
# Creating New tags column with combining overview, genres, keywords, cast, crew.
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew'] 
```

```python
movies.head()
```

```python
# Creating Final Data for selected Columns
final_data = movies[["movie_id", "title", "tags"]]
```

```python
final_data['tags'] = final_data['tags'].apply(lambda x: " ".join(x))
```

```python
final_data['tags'] = final_data['tags'].apply(lambda x: x.lower())
```

```python
final_data.head()
```

```python
final_data.to_csv('final_data.csv', index=False)
```
